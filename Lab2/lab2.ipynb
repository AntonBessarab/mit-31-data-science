```python
import matplotlib.pyplot as plt
import csv

x_true = np.arange(0, 20, 2)
y_true = 0.5 * x_true - 2
y_pred = np.array([-1.5, -1, 0, 0.5, 3.15, 3, 3.5, 4, 4.1, 6])

mae = np.mean(np.abs(y_true - y_pred))
mse = np.mean((y_true - y_pred)**2)
print(f'MAE: {mae}')
print(f'MSE: {mse}')

plt.plot(x_true, y_true, label="Linear function y=0,5x-2")
plt.plot(x_true, y_pred, label="Predicted values")
plt.title("Line graph")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()


with open("results.csv", mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Metric", "Value"])
    writer.writerow([f"MAE  , {mae}"])
    writer.writerow([f"MSE  , {mse}"])
```
