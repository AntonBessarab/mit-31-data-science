Створення набору даних у вигляді тензорів

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Скаляр
scalar = tf.constant(42)
print("Скаляр:", scalar.numpy())

# Вектор
vector = tf.constant([1, 2, 3, 4, 5])
print("Вектор:", vector.numpy())

# Матриця
matrix = tf.constant([[1, 2, 3], [4, 5, 6]])
print("Матриця:")
print(matrix.numpy())
```

CRUD операції

```python
# Читання елементів
print("Елемент матриці [0, 1]:", matrix[0, 1].numpy())
print("Сегмент вектора [1:4]:", vector[1:4].numpy())

# Оновлення матриці
matrix_updated = tf.tensor_scatter_nd_update(matrix, [[0, 1]], [999])
print("Оновлена матриця:")
print(matrix_updated.numpy())

# Додавання елементів
vector_updated = tf.concat([vector, [6, 7]], axis=0)
print("Вектор після додавання елементів:", vector_updated.numpy())

# Видалення елементів
vector_reduced = tf.boolean_mask(vector_updated, [True, True, False, True, True, True, False])
print("Вектор після видалення елементів:", vector_reduced.numpy())
```

Генерація даних для регресії та класифікації

```python
# Регресія
x_reg = tf.linspace(-1.0, 1.0, 100)
y_reg = 3 * x_reg + tf.random.normal([100], mean=0.0, stddev=0.2)


# Класифікація
n_samples = 200
class_0 = tf.random.normal([n_samples // 2, 2], mean=0.0, stddev=1.0)
class_1 = tf.random.normal([n_samples // 2, 2], mean=3.0, stddev=1.0)
x_class = tf.concat([class_0, class_1], axis=0)
y_class = tf.concat([tf.zeros([n_samples // 2]), tf.ones([n_samples // 2])], axis=0)
```

Графічний аналіз даних

```python
# Візуалізація даних регресії
plt.figure(figsize=(10, 5))
plt.scatter(x_reg, y_reg, label="Дані регресії")
plt.plot(x_reg, 3 * x_reg, color='red', label="Істинна функція")
plt.title("Дані для регресії")
plt.legend()
plt.show()

# Візуалізація даних класифікації
plt.figure(figsize=(10, 5))
plt.scatter(class_0[:, 0], class_0[:, 1], label="Клас 0", alpha=0.7)
plt.scatter(class_1[:, 0], class_1[:, 1], label="Клас 1", alpha=0.7)
plt.title("Дані для класифікації")
plt.legend()
plt.show()
```
